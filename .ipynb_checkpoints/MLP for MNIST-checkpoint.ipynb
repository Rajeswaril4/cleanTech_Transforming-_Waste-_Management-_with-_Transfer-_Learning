# Import the libraries

```python
# importing modules
import tensorflow as tf
import numpy as np
from keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Activation
import matplotlib.pyplot as plt
```

# Load the dataset and split the dataset

```python
(x_train, y_train), (x_test, y_test) =mnist.load_data()
```

# displaying the first 20 images of the training dataset

```python
for i in range(20):
    plt.subplot(5,4,i+1)
    plt.imshow(x_train[i],cmap='gray_r')
    plt.title("Digits:{}".format(y_train[i]))
    plt.subplots_adjust(hspace=0.5)
    plt.axis('off')
```

```python
plt.hist(x_train[0].reshape(784),facecolor='green')
plt.title('pixel vs its intensity',fontsize=20)
plt.ylabel('pixel')
plt.xlabel('intensity')
plt.text(0.5,0,'intensity')
```

# converting pixels into floating values

```python
# Cast the records into float values
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
```

# normalize the training and testing data

```python
gray_scale = 255
x_train /= gray_scale
x_test /= gray_scale
```

# print the shape of the splitted data

```python
print("Feature matrix:", x_train.shape)
print("Target matrix:", x_test.shape)
print("Feature matrix:", y_train.shape)
print("Target matrix:", y_test.shape)
```

# creating sequential model

```python
model = Sequential([
	
	# reshape 28 row * 28 column data to 28*28 rows
	Flatten(input_shape=(28, 28)),
	
	# dense layer 1
	Dense(256, activation='sigmoid'),
	
	# dense layer 2
	Dense(128, activation='sigmoid'),
	
	# output layer
	Dense(10, activation='sigmoid'),
])
```

# compile the model

```python
model.compile(optimizer='adam',
			loss='sparse_categorical_crossentropy',
			metrics=['accuracy'])
```

# fit the model

```python
model.fit(x_train, y_train, epochs=10,
		batch_size=2000,
		validation_split=0.2)
```

# evaluate the model

```python
results = model.evaluate(x_test, y_test, verbose = 0)
print('test loss, test acc:', results)
```
